```python
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get('https://wuzzuf.net/search/jobs/?q=Machine%20learning&a=hpb%27')
src = page.content
src
```

```python
soup = BeautifulSoup(src, 'lxml')
```

```python
jobs =soup.find_all("div",class_="css-1gatmva e1v1l3u10")
```

```python
import pandas as pd
df = pd.DataFrame(columns = ["job_title", "company_name","area","city","country","job_type" ,"job_state","exp_level", "exp_years","job_category","skills","posted"])
df
```

```python
import numpy as np

# Example extracted text
exp_text = jobs[0].find_all("div", class_ = "css-1lh32fc")[0].next_sibling.text.split(" · ")[1]

def parse_exp_years(exp_text):
    try:
        # Extract years from the text
        years = exp_text.split(" Yrs of Exp")[0]

        # Check if the years part contains a range
        if '-' in years:
            start_year, end_year = years.split('-')
            start_year = start_year.strip()
            end_year = end_year.strip()

            # Check if both start and end year are valid numbers
            if start_year.isdigit() and end_year.isdigit():
                return exp_text
            else:
                return np.nan
        else:
            # Check if the single year value is a valid number
            if years.strip().isdigit():
                return exp_text
            else:
                return np.nan
    except Exception as e:
        return np.nan

exp_years = parse_exp_years(exp_text)
print(exp_years)
```

```python
#Machine Learning jobs
#iterating on every page
for i in range(68):
    pages = requests.get(f"https://wuzzuf.net/search/jobs/?a=hpb%7Cspbg&q=machine%20learning&start={i}")
    src = pages.content
    soup = BeautifulSoup(src, 'lxml')
    jobs =soup.find_all("div",class_="css-1gatmva e1v1l3u10")
    for job in jobs:
        job_title=job.find_all("h2",class_="css-m604qf")[0].text
        company_name = job.find_all("div", class_ = "css-d7j1kk")[0].text.split(" - ")[0]
        try:

            area = job.find_all("span", class_ = "css-5wys0k")[0].text.split(", ")[-3]

        except:
            area="Not mentioned"
        city = job.find_all("span", class_ = "css-5wys0k")[0].text.split(", ")[-2]
        country = job.find_all("span", class_ = "css-5wys0k")[0].text.split(", ")[-1]
        try:
            job_state = jobs[1].find_all("span", class_ = "css-o1vzmt eoyjyou0")[0].text

        except:
             job_state = "Not Mentioned"
        job_type = jobs[0].find_all("span", class_ = "css-1ve4b75 eoyjyou0")[0].text
        exp_level = job.find_all("div", class_ = "css-1lh32fc")[0].next_sibling.text.split(" · ")[0]
        try:
            posted = job.find_all("div", class_ = "css-do6t5g")[0].text
        except:
            posted = job.find_all("div", class_ = "css-4c4ojb")[0].text
        job_category= job.find_all("div", class_ = "css-1lh32fc")[0].next_sibling.text.split(" · ")[2]
        skills = " ".join(job.find_all("div", class_ = "css-1lh32fc")[0].next_sibling.text.split(" · ")[5:])
        exp_text = job.find_all("div", class_ = "css-1lh32fc")[0].next_sibling.text.split(" · ")[1]
        exp_years = parse_exp_years(exp_text)


        df.loc[len(df.index)] =[job_title, company_name,area,city,country,job_type ,job_state,exp_level,exp_years,job_category,skills,posted]
```
